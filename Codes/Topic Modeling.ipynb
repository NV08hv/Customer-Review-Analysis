```python
import numpy as np
import pandas as pd
import re
import os
import sys
import nltk
import warnings
from gensim.models import CoherenceModel
warnings.filterwarnings(action='ignore',category=DeprecationWarning)
warnings.filterwarnings(action='ignore',category=FutureWarning)
pd.options.display.max_columns = None
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer
```

### Read Dataset

```python
data = pd.read_csv('/content/drive/MyDrive/Course DA/Final Project/Final/data/customer_review.csv')
data.head()
```

Divide dataset into 2 part:


*   Positive Review: rating >= 3
*   Negative Review: rating < 3



### Positive review

```python
df_positive = data[data['rating'] >= 3]
df_positive.head()
```

```python
def minimal_cleaning(text):
    # Remove URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    # Remove emoji, signal
    text = re.sub(r'[^A-Za-z0-9\s.,!?\'\"]', '', text)

    # Remove many space
    text = re.sub(r'\s+', ' ', text).strip()

    return text

df_positive['review_content'] = df_positive['review_content'].astype(str).apply(minimal_cleaning)
docs = df_positive['review_content'].to_list()
docs
```

#### Bertopic

```python
from gensim.corpora import Dictionary
docs_tokens = [nltk.word_tokenize(doc.lower()) for doc in docs]
dictionary = Dictionary(docs_tokens)
```

```python
for k in [2, 3, 4, 5,10,15,20,25]:
    vectorizer_model = CountVectorizer(stop_words='english')
    model = BERTopic(nr_topics=k,vectorizer_model=vectorizer_model,
                     calculate_probabilities=False)
    topics, _ = model.fit_transform(docs)
    topics = model.get_topics()

    topic_terms = [
    [term for term, _ in topics[topic_id]]
    for topic_id in topics
    if topic_id != -1]

    coherence_scores = []
    for terms in topic_terms:
        cm = CoherenceModel(
            topics=[terms],
            texts=docs_tokens,
            dictionary=dictionary,
            coherence='c_v'
        )
        coherence_scores.append(cm.get_coherence())

    mean_coherence = np.mean(coherence_scores)
    print(f"With k = {k} Mean topic coherence (c_v): {mean_coherence:.4f}")
```

```python
vectorizer_model = CountVectorizer(stop_words='english')
model = BERTopic(nr_topics=5, verbose=True, vectorizer_model=vectorizer_model)
model.fit(docs)
topics, probabilities = model.transform(docs)
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import ast

load_dotenv()
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))

def create_gpt_prompts(df):
  prompts = []

  for i, row in df.iterrows():
    topic = row['Topic']
    name = row['Name']
    keywords = ' '.join(ast.literal_eval(row['Representation']))
    examples = ast.literal_eval(row['Representative_Docs'])[:3]
    prompt = (
            f"Topic số {topic}: {name}, \n"
            f"Các từ khóa nổi bật: {keywords}, \n"
            f"Một vài ví dụ tiêu biểu từ người dùng:, \n{examples} \n"
            f"Hãy viết một đoạn nhận xét hoặc mô tả tổng quát về chủ đề này."
        )
    prompts.append(prompt)
  return prompts
```

```python
create_gpt_prompts(info_negative_topic.head(1))
```

### GPT generation With negative reviews

```python
for prompt in create_gpt_prompts(info_negative_topic):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Bạn là một nhà phân tích phản hồi khách hàng."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.7,
        max_tokens=200
    )
    print("Prompt:\n", prompt)
    print("GPT trả lời:\n", response.choices[0].message.content)
    print("\n" + "="*80 + "\n")
```
